Реализовать следующие структуры данных:
  
● <b> Стек (stack): </b>  
Операции для стека:
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, извлечение элемента из начала;  </i>  
  
● <b> Дек (двусторонняя очередь, deque):  </b>  
Операции для дека:
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, добавление нового элемента в конец, извлечение элемента из начала, извлечение элемента из конца. </i>  
  
<i> Разработать программу обработки данных, содержащихся в заранее подготовленном txt-файле, в соответствии с заданиями, применив указанную в задании структуру данных. Результат работы программы вывести на экран и сохранить в отдельном txt-файле. </i>

```python
# Отсортировать строки файла, содержащие названия книг, в алфавитном порядке с использованием двух деков.
class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return None
        return self.stack.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.stack[-1]

class Deque:
    def __init__(self):
        self.deque = []

    def is_empty(self):
        return len(self.deque) == 0

    def add_front(self, item):
        self.deque.insert(0, item)

    def add_rear(self, item):
        self.deque.append(item)

    def remove_front(self):
        if self.is_empty():
            return None
        return self.deque.pop(0)

    def remove_rear(self):
        if self.is_empty():
            return None
        return self.deque.pop()

# Функция для сортировки строк с использованием двух деков
def sort_books(input_file, output_file):
    deque1 = Deque()
    deque2 = Deque()
    temp_deque = Deque()  # Добавлен временный deque

    with open(input_file, 'r', encoding='utf-8') as file:
        lines = [line.strip() for line in file.readlines() if line.strip()]

    for line in lines:
        deque1.add_rear(line)

    while not deque1.is_empty():
        min_value = deque1.remove_front()
        while not deque1.is_empty():
            item = deque1.remove_front()
            if item < min_value:
                temp_deque.add_rear(min_value)  # Используем temp_deque
                min_value = item
            else:
                temp_deque.add_rear(item) # Используем temp_deque

        deque2.add_rear(min_value)
        # Перемещаем элементы из temp_deque обратно в deque1
        while not temp_deque.is_empty():
            deque1.add_rear(temp_deque.remove_front())


    sorted_books = []
    while not deque2.is_empty():
        sorted_books.append(deque2.remove_front())

    # Запись результата в файл
    with open(output_file, 'w', encoding='utf-8') as file:
        for book in sorted_books:
            file.write(book + '\n')

    # Вывод результата в консоль
    print("--- Отсортированные книги ---")
    for book in sorted_books:
        print(book)

# Файлы ввода и вывода
input_filename = "input.txt"
output_filename = "output.txt"

# Запуск сортировки
try:
    sort_books(input_filename, output_filename)
    print(f"Сортировка завершена. Результат сохранён в {output_filename}")
except FileNotFoundError:
    print(f"Ошибка: файл {input_filename} не найден.")
```

```python
# Дек содержит последовательность символов для шифровки сообщений. 
# Дан текстовый файл, содержащий зашифрованное сообщение.    
# Известно, что при шифровке каждый символ сообщения заменялся следующим за ним в деке по часовой стрелке через один.
class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return None
        return self.stack.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.stack[-1]

class Deque:
    def __init__(self):
        self.deque = []

    def is_empty(self):
        return len(self.deque) == 0

    def add_front(self, item):
        self.deque.insert(0, item)

    def add_rear(self, item):
        self.deque.append(item)

    def remove_front(self):
        if self.is_empty():
            return None
        return self.deque.pop(0)

    def remove_rear(self):
        if self.is_empty():
            return None
        return self.deque.pop()

    def rotate_clockwise(self, steps=1):
        for _ in range(steps):
            if not self.is_empty():
                self.deque.insert(0, self.deque.pop())

    def rotate_counterclockwise(self, steps=1):
        for _ in range(steps):
            if not self.is_empty():
                self.deque.append(self.deque.pop(0))

# Функция для расшифровки сообщения
def decrypt_message(input_file, output_file, sequence):
    deque = Deque()
    for char in sequence:
        deque.add_rear(char)

    with open(input_file, 'r', encoding='utf-8') as file:
        encrypted_text = file.read().strip()

    decrypted_text = ""
    for char in encrypted_text:
        if char in deque.deque:
            index = deque.deque.index(char)
            deque.rotate_counterclockwise(index + 2)  # Смещение в обратную сторону
            decrypted_text += deque.deque[0]
            deque.rotate_clockwise(index + 2)  # Возвращаем исходный порядок
        else:
            decrypted_text += char

    # Запись расшифрованного текста в файл
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(decrypted_text)

    # Вывод расшифрованного текста в консоль
    print("--- Расшифрованное сообщение ---")
    print(decrypted_text)

# Файлы ввода и вывода
input_filename = "encrypted.txt"
output_filename = "decrypted.txt"
char_sequence = "АБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯабвгдеёжзийклмнопрстуфхцчшщъыьэюяABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789.,!? "  # Пример последовательности символов

# Расшифровка сообщения
try:
    decrypt_message(input_filename, output_filename, char_sequence)
    print(f"Расшифровка завершена. Результат сохранён в {output_filename}")
except FileNotFoundError:
    print(f"Ошибка: файл {input_filename} не найден.")
```

```python
# Задание №3
#Даны три стержня и n дисков различного размера. Диски можно надевать на стержни, образуя из них башни. Перенести n дисков со стержня А на стержень С, сохранив их первоначальный порядок. При переносе дисков необходимо соблюдать следующие правила:
#- на каждом шаге со стержня на стержень переносить только один диск;
#- диск нельзя помещать на диск меньшего размера;
#- для промежуточного хранения можно использовать стержень В.
#Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return None
        return self.stack.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.stack[-1]

# Функция перемещения дисков
def move_disks(n, source, target, auxiliary):
    if n > 0:
        move_disks(n - 1, source, auxiliary, target)
        disk = source.pop()
        if disk is not None:
            target.push(disk)
            print(f"Переместить диск {disk} со стержня {source.name} на {target.name}")
        move_disks(n - 1, auxiliary, target, source)

# Функция для чтения данных и решения задачи Ханойской башни
def solve_hanoi(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as file:
        disks = [int(line.strip()) for line in file.readlines() if line.strip().isdigit()]

    source = Stack()
    target = Stack()
    auxiliary = Stack()

    source.name = "A"
    target.name = "C"
    auxiliary.name = "B"

    for disk in sorted(disks, reverse=True):
        source.push(disk)

    print("--- Начало перемещения ---")
    move_disks(len(disks), source, target, auxiliary)
    print("--- Перемещение завершено ---")

    # Запись результата в файл
    with open(output_file, 'w', encoding='utf-8') as file:
        while not target.is_empty():
            file.write(str(target.pop()) + '\n')
    print(f"Результат сохранён в {output_file}")

# Файлы ввода и вывода
input_filename = "input2.txt"
output_filename = "output2.txt"

# Решение задачи Ханойской башни
try:
    solve_hanoi(input_filename, output_filename)
except FileNotFoundError:
    print(f"Ошибка: файл {input_filename} не найден.")
```

```python
#Дан текстовый файл с программой на алгоритмическом языке. 
# За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.
class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return None
        return self.stack.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.stack[-1]

# Функция проверки баланса круглых скобок
def check_parentheses_balance(input_file):
    stack = Stack()

    with open(input_file, 'r', encoding='utf-8') as file:
        text = file.read()

    for char in text:
        if char == '(':
            stack.push(char)
        elif char == ')':
            if stack.is_empty():
                return False  # Закрывающая скобка без открывающей
            stack.pop()

    return stack.is_empty()  # Если стек пустой, баланс соблюден

# Файлы ввода
input_filename = "program.txt"

# Проверка баланса скобок
try:
    if check_parentheses_balance(input_filename):
        print("Баланс скобок соблюден.")
    else:
        print("Ошибка: несбалансированные скобки.")
except FileNotFoundError:
    print(f"Ошибка: файл {input_filename} не найден.")
```

```python
# Дан файл из символов. Используя стек, за один просмотр файла напечатать сначала все цифры, 
# затем все буквы, и, наконец, все остальные символы, сохраняя исходный порядок в каждой группе символов.
class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return None
        return self.stack.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.stack[-1]

# Функция для разделения символов и печати в заданном порядке
def process_file_by_character_type(input_file):
    digits_stack = Stack()
    letters_stack = Stack()
    others_stack = Stack()

    with open(input_file, 'r', encoding='utf-8') as file:
        text = file.read()

    digits = []
    letters = []
    others = []

    for char in text:
        if char.isdigit():
            digits.append(char)
        elif char.isalpha():
            letters.append(char)
        else:
            others.append(char)

    for char in digits:
        digits_stack.push(char)
    for char in letters:
        letters_stack.push(char)
    for char in others:
        others_stack.push(char)

    result = "".join(digits + letters + others)

    print("--- Отсортированные символы ---")
    print(result)

    return result

# Файлы ввода
input_filename = "characters.txt"

# Обработка файла
try:
    sorted_result = process_file_by_character_type(input_filename)
    with open("sorted_output.txt", 'w', encoding='utf-8') as file:
        file.write(sorted_result)
    print("Результат сохранен в sorted_output.txt")
except FileNotFoundError:
    print(f"Ошибка: файл {input_filename} не найден.")
```

```python
# Дан файл из целых чисел. Используя дек, за один просмотр файла напечатать сначала все отрицательные числа, 
# затем все положительные числа, сохраняя исходный порядок в каждой группе.
class Deque:
    def __init__(self):
        self.deque = []

    def is_empty(self):
        return len(self.deque) == 0

    def add_front(self, item):
        self.deque.insert(0, item)

    def add_rear(self, item):
        self.deque.append(item)

    def remove_front(self):
        if self.is_empty():
            return None
        return self.deque.pop(0)

    def remove_rear(self):
        if self.is_empty():
            return None
        return self.deque.pop()

# Функция для сортировки чисел по знаку
def process_numbers_by_sign(input_file):
    negative_deque = Deque()
    positive_deque = Deque()

    with open(input_file, 'r', encoding='utf-8') as file:
        numbers = [int(num) for num in file.read().split() if num.lstrip('-').isdigit()]

    for num in numbers:
        if num < 0:
            negative_deque.add_rear(num)
        else:
            positive_deque.add_rear(num)

    result = [str(negative_deque.remove_front()) for _ in range(len(negative_deque.deque))] + \
             [str(positive_deque.remove_front()) for _ in range(len(positive_deque.deque))]

    result_text = " ".join(result)

    print("--- Отсортированные числа ---")
    print(result_text)

    return result_text

# Файлы ввода
input_filename = "numbers.txt"

# Обработка файла
try:
    sorted_result = process_numbers_by_sign(input_filename)
    with open("sorted_numbers_output.txt", 'w', encoding='utf-8') as file:
        file.write(sorted_result)
    print("Результат сохранен в sorted_numbers_output.txt")
except FileNotFoundError:
    print(f"Ошибка: файл {input_filename} не найден.")
```

```python
# Дан текстовый файл. Используя стек, сформировать новый текстовый файл, 
# содержащий строки исходного файла, записанные в обратном порядке: 
# первая строка становится последней, вторая – предпоследней и т.д.
class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return None
        return self.stack.pop()

# Функция для переворачивания строк файла
def reverse_file_lines(input_file, output_file):
    stack = Stack()

    with open(input_file, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    for line in lines:
        stack.push(line.strip())

    with open(output_file, 'w', encoding='utf-8') as file:
        while not stack.is_empty():
            file.write(stack.pop() + '\n')

    print(f"Файл {output_file} создан с перевернутыми строками.")

# Файлы ввода и вывода
input_filename = "input3.txt"
output_filename = "output3.txt"

# Запуск переворота строк
try:
    reverse_file_lines(input_filename, output_filename)
except FileNotFoundError:
    print(f"Ошибка: файл {input_filename} не найден.")
```
